```python
# Downlaod the Data with url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
tables = pd.read_html(url)
df_toronto = tables[0]
df_toronto.head()
```

```python
df_toronto = df_toronto[df_toronto.Borough != 'Not assigned']
df_toronto
```

```python
# simply drop whole row with NaN in "borough"
df_toronto.dropna(subset=["Borough"], axis=0, inplace=True)
df_toronto
```

```python
# Groupby Borough
df_toronto = df_toronto.groupby(['Postcode','Borough'], sort = True).agg(lambda x: ', '.join(x))
df_toronto
```

```python
df_toronto.reset_index(inplace = True)
df_toronto
```

```python
# Rename Postcode to Postal Code
df_toronto.rename(columns={'Postcode':'Postal Code'}, inplace=True)
df_toronto
```

```python
# Change Not assigned Neighbourhood to Borough name, e.g Queen's Park
df_toronto.loc[df_toronto['Neighbourhood']=="Not assigned",'Neighbourhood']=df_toronto.loc[df_toronto['Neighbourhood']=="Not assigned",'Borough']
df_toronto
```

```python
# Read Geospatial Data: Latitudes and Longitudes of Toronto Neighbourhoods
df_toronto_ll = pd.read_csv('http://cocl.us/Geospatial_data')
df_toronto_ll.head(5)
```

```python
# Merge Files on Postal Code to get consolidated data on Postal Code Boroughs, Neighbourhoods and their Latitudes and Longitudes

df_torontofull = pd.merge(df_toronto, df_toronto_ll, on='Postal Code')
df_torontofull
```

```python
# Latitude and Logitude of Toronto, CA

address = 'Toronto, CA'

geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# define the Toronto map
toronto_map = folium.Map(location=[43.653963, -79.387207], zoom_start=10)

# display Toronto Map
toronto_map
```

```python
# add markers to map
for lat, lng, borough, neighbourhood in zip(df_torontofull['Latitude'], df_torontofull['Longitude'], df_torontofull['Borough'], df_torontofull['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(toronto_map)  
    
toronto_map
```

```python
# Extracting data with Borough's containing Toronto
df_onlyToronto = df_torontofull[df_torontofull['Borough'].str.contains('Toronto',regex=False)]
df_onlyToronto
```

```python
toronto_map = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df_onlyToronto['Latitude'],df_onlyToronto['Longitude'],df_onlyToronto['Borough'],df_onlyToronto['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(toronto_map)
toronto_map
```

```python
# Creating different clusters for East Toronto, West Toronto, Central Toronto and Downtown Toronto

k=5
toronto_clustering = df_onlyToronto.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df_onlyToronto.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
df_onlyToronto
```

```python
# create map with clusters
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=11)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df_onlyToronto['Latitude'], df_onlyToronto['Longitude'], df_onlyToronto['Neighbourhood'], df_onlyToronto['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
